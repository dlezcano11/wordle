COMO FUNCIONA EL WORDLE.
0. Se define la palabra secreta de la partida en el programa
1. Avisa la cantidad de intentos, en este caso son 6. 
2. Te solicita la primera palabra. Esta debe ser de 5 letras. 
    Se puede hacer una funcion verificadora para que se cumpla este punto.
3. Una vez ingresada una palabra valida, esta se compara con la palabra secreta.
    Si coinciden letra y posicion, ES VERDE. (Se imprime la letra entre llaves)
    Si coincide letra pero no posicion, ES AMARILLO. (Se imprime la letra entre parentesis)
    Si no coincide ni letra ni posicion, ES ROJO. (Se imprime la letra sin nada)

    OBS: Si todas las letras de la misma partida son VERDE, termina el juego. Imprimir mensaje de felicitacion.
    Se puede usar un ciclo while, que debe depender de la cantidad de intentos. 

    Si se usan todos los intentos y no se adivina la palabra, se debe terminar el juego con un mensaje.
